```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import Normalize
from matplotlib import cm
import mapclassify as mc
from matplotlib_scalebar.scalebar import ScaleBar

os.chdir('C:\\Users\\tuo27112\\OneDrive - Temple University\\Geoviz\\Labs\\final\\data\\')

# read the data
heat_shp = r'heat_exposure.shp'
heat_gdf = gpd.read_file(heat_shp)


# change the projection to local projection of Philadelphia
heat_philly = heat_gdf.to_crs(epsg=2272) 
list(heat_gdf.columns)
```

```python

# create a empty plot for the choropleth map
fig, ax = plt.subplots(1, figsize=(8, 8))

# the number of categories
n_class = 5

# get the color bar, the string for color can be found here, https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html
cmap = plt.cm.get_cmap('bwr', n_class)


# the field to plot the choropleth map
field = 'AvgTempDif'

# plot the choropleth map of blood lead in Philadelphia
heat_philly.plot(column=field, 
                     cmap='bwr', 
                     edgecolor='0.5', 
                     ax = ax,
                     linewidth=0.5,
#                      legend=True, #don't use the legend
                     k=n_class, #the number of classes
                     scheme='NaturalBreaks') # 'equal_interval', 'quantiles', 'FisherJenks', 'NaturalBreaks'


# the function should be matched with the scheme of the above geopandas plot, https://pysal.org/mapclassify/_modules/mapclassify/classifiers.html
nb = mc.NaturalBreaks(heat_gdf[field].dropna(), k=n_class) # NaturalBreaks, FisherJenks, quantile
x = list(nb.bins) #get the interval labels
x.insert(0, heat_gdf[field].min())


# set the location of the legend
axins = ax.inset_axes([1, 0.2, 0.03, 0.6]) #x0, y0, width, height

# # create a color bar for the map
# norm = Normalize(vmin=blood_lead_gdf[field].min(), vmax=blood_lead_gdf[field].max(), clip=True)
norm = mpl.colors.BoundaryNorm(vals, cmap.N) #discrete color bar
n_cmap = cm.ScalarMappable(norm=norm, cmap=cmap)
font3 = {'family' : 'Open Sans',
        'weight' : 'regular',
        'size'   : 12}
mpl.rc('font', **font3)
cbar = ax.get_figure().colorbar(n_cmap, 
                                orientation='vertical', 
                                cax=axins,
                                spacing='proportional',  #let the distance in legend proportional to the value
                                fraction=0.046, 
                                pad=0.0, 
                                ticks=x,
                                shrink=0.4)

font = {'family' : 'Open Sans',
        'weight' : 'bold',
        'size'   : 16}

mpl.rc('font', **font)
ax.set_title('Average Temperature Difference ($^\circ$F) from City Average by Census Tract')

# add scale bar
# Position and layout
font2 = {'family' : 'Open Sans',
        'weight' : 'regular',
        'size'   : 10}
mpl.rc('font', **font2)
scale1 = ScaleBar(
    dx=1, 
    location='lower right',  # in relation to the whole plot
    label_loc='left', scale_loc='bottom'  # in relation to the line
    
)

# trying to add degree symbols ($^\circ$) to color bar labels
pos = 'both'
ax.xaxis.set_ticks_position(pos)
fmtr = mpl.ticker.StrMethodFormatter('{x:} $^\circ$')
ax.yaxis.set_major_formatter(fmtr)
ax.add_artist(scale1)


ax.set_axis_off()
plt.axis('equal')
#plt.savefig("phl_blood_lead.png", dpi=200)
plt.show()
```
